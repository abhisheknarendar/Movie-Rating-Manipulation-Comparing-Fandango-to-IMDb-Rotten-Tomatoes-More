                                   SPLIT THE YEAR FROM THE TITLE AND DISPLAY IT AS SEPERATE ROW

```python
fandango.head()
```

```python
def get_year(title):
    x = title.split('(')[-1].replace(")",'')
    return x
    
```

```python
fandango['YEAR'] = fandango['FILM'].apply(get_year)
```

                                                  FANDANGO SCORES VS ALL SITES

```python
combined = pd.merge(reviewed,all_sites,on = "FILM")
```

```python
combined.head()
```

```python
combined.info()
```

```python
combined.describe().transpose()
```

```python
#NORMALIZE ALL COLUMNS SO THE RATINGS FOR OTHER SITES MATCH TO THE 0-5 RANGE STARS IN FANDANGO
combined["RT_NORM"] = np.round(combined["RottenTomatoes"] / 20, 1)
combined["RTUSER_NORM"] = np.round(combined["RottenTomatoes_User"] / 20, 1)
combined["META_NORM"] = np.round(combined["Metacritic"] / 20, 1)
combined["METAUSER_NORM"] = np.round(combined["Metacritic_User"] / 2, 1)
combined["IMDB_NORM"] = np.round(combined["IMDB"] / 2, 1)
```

```python
combined.head()
```

```python
combined.columns
```

```python
#CREATE A TALE WITH ONLY STARS , RATINGS AND THE NORMALIZED COLUMNS
normalized_final = combined[['STARS', 'RATING','RT_NORM', 'RTUSER_NORM',
       'META_NORM', 'METAUSER_NORM', 'IMDB_NORM']]
```

```python
normalized_final
```

                                             VISUALIZING THE DIFFERENCE AMONG ALL THE SITES 

```python
plt.figure(figsize = (10,4),dpi = 200)
sns.kdeplot(data = normalized_final,clip = [0,5],fill = True)
```

     THE ROTTEN TOMATOES CRITICS SEEM TO HAVE THE MOST UNIFORM DISTRIBUTION SO LET'S COMPARE ROTTEN TOMATOES CRITICS VS FANDANGO STARS

```python
plt.figure(figsize = (10,4),dpi = 200)
sns.kdeplot(data = normalized_final, x = "RT_NORM",clip = [0,5],fill = True)
sns.kdeplot(data = normalized_final, x = "STARS",clip = [0,5],fill = True)
```

```python
sns.clustermap(data = normalized_final)
```

CLEARLY HERE FANDANGO DOESNT HAVE ANY RATINGS UNDER 3 ATMOST WHICH SHOWS HOW MUCH IT INCREASES RATINGS FOR MOVIES WHICH HAVE BEEN GIVEN LOW SCORES BY OTHER SITES

```python
#GRAB THE TOP 10 WORST FILMS AND VISUALIZE IT
normalized = combined[['FILM','STARS', 'RATING','RT_NORM', 'RTUSER_NORM',
       'META_NORM', 'METAUSER_NORM', 'IMDB_NORM']]
normalized.nsmallest(10,"RT_NORM")
```

```python
worst_10 = normalized.nsmallest(10,"RT_NORM")
```
